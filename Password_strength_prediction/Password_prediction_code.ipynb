```python
df_array=np.array(df)
df_array
```

```python
import random
random.shuffle(df_array)
```

```python
x=[label[0] for label in df_array]
y=[label[1] for label in df_array]
```

```python
x
```

```python
y
```

```python
sns.countplot(y)
```

```python
def alphabet_splitter(str):
    lst=[]
    for i in str:
        lst.append(i)
    return lst
```

```python
alphabet_splitter('abcdefghij')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vector_converter=TfidfVectorizer(tokenizer=alphabet_splitter)
```

```python
X=vector_converter.fit_transform(x)
```

```python
X.shape
```

```python
vector_converter.get_feature_names()
```

```python
X
```

```python
first_password_vector=X[0]
first_password_vector
```

```python
first_password_vector.T
```

```python
first_password_vector.T.todense()
```

```python
first_password_dataframe=pd.DataFrame(first_password_vector.T.todense(), index=vector_converter.get_feature_names(), columns=["TF-IDF"])
```

```python
first_password_dataframe.sort_values(by=['TF-IDF'],ascending=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression(random_state=45, multi_class='multinomial')
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm_logistic=confusion_matrix(y_test,y_pred)
print(cm_logistic)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_classifier=DecisionTreeClassifier(criterion='entropy', splitter='best', random_state = 100, max_depth=40, min_samples_leaf=5)
```

```python
tree_classifier.fit(X_train,y_train)
```

```python
y_pred=tree_classifier.predict(X_test)
y_pred
```

```python
cm_decision_tree=confusion_matrix(y_test,y_pred)
print(cm_decision_tree)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest_classifier=RandomForestClassifier(n_estimators=100, random_state = 100, max_depth=50, criterion='entropy', min_samples_leaf=5)
```

```python
forest_classifier.fit(X_train,y_train)
```

```python
y_pred=forest_classifier.predict(X_test)
y_pred
```

```python
cm_random_forest=confusion_matrix(y_test,y_pred)
print(cm_random_forest)
print(accuracy_score(y_test,y_pred))
```
