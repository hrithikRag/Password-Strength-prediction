```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Define the number of classes and input dimension
num_classes = 3
input_dim= X_train.shape[1]
# Create a sequential model
model = Sequential()

# Adding the input layer
model.add(Dense(units=100, activation='relu', input_dim=input_dim))  # Replace input_dim with your input dimension

# Adding more hidden layers
#model.add(tf.keras.layers.Dropout(0.2))
model.add(Dense(units=300, activation='relu'))
model.add(Dense(units=200, activation='relu'))
model.add(Dense(units=100, activation='relu'))
# Adding the output layer with softmax activation for multi-class classification
model.add(Dense(units=num_classes, activation='softmax'))

# Define the Adam optimizer and specify the learning rate
adam_optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

# Compile the model
model.compile(optimizer=adam_optimizer, loss='sparse_categorical_crossentropy', metrics=['sparse_categorical_accuracy'])
model.summary()
```

```python
model.fit(X_train, y_train, batch_size=500, epochs=100)
```

```python
adam_optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)    #further optimizing by decreasing the learning rate
```

```python
model.fit(X_train, y_train, batch_size=500, epochs=50)
```

```python
test_loss, test_accuracy= model.evaluate(X_test, y_test)
print(test_accuracy)
```

```python
output_values= model.predict(X_test)
y_pred= np.argmax(output_values, axis=1)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)
print(cm)
```
